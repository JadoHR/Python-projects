write a definition for a class named circle with attributes center and radius, where a center is a point object and a radius is a number

```python
class Circle:
    def __init__(self, center, radius):
        self.center = center
        self.radius = radius
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
circle = Circle(Point(0, 0), 5)
```

- Write a function named ptInCircle that takes a Circle and a Point and returns True if the Point lies in or on the boundary of the circle.

```python
def ptInCircle(circle, point):
    distance = ((point.x - circle.center.x) ** 2 +
                (point.y - circle.center.y) ** 2) ** 0.5
    return distance <= circle.radius
```

```python
circle = Circle(Point(0, 0), 5)
point = Point(3, 4)

print(ptInCircle(circle, point))
```

```python
import math
def distBtwnPoints(p1, p2):
    """Computes the distance between two Point objects.
    p1: Point
    p2: Point
    returns: float
    """
    dx = p1.x - p2.x
    dy = p1.y - p2.y
    dist = math.sqrt(dx**2 + dy**2)
    return dist
```

```python
def ptInCircle1(point, circle):
    """Checks whether a point lies inside a circle (or on the boundary).
    point: Point object
    circle: Circle object
    """
    d = distBtwnPoints(point, circle.center)
    print(d)
    return d <= circle.radius
```

```python

import numpy as np

# Generate random 2D array (3x4)
array = np.random.randint(1, 10, size=(3, 4))

# Calculate the mean of the array
mean = np.mean(array)

# Replace values greater than mean with 0
array[array > mean] = 0

# Generate another random 2D array (2x3)
another_array = np.random.randint(1, 10, size=(2, 3))

# Multiply the two arrays
result = np.dot(array, another_array)

print("Modified Array:")
print(array)
print("\nAnother Array:")
print(another_array)
print("\nResult of Multiplication:")
print(result)
```

```python
import numpy as np

# Create a 3x4 matrix of random integers between 0 and 99
matrix1 = np.random.randint(0, 100, size=(3, 4))
print("Matrix 1:")
print(matrix1)

# Calculate the mean of the matrix
mean = np.mean(matrix1)
print("\nMean:", mean)

# Replace all values greater than the mean with 0
matrix1 = np.where(matrix1 > mean, 0, matrix1)
print("\nMatrix 1 (Modified):")
print(matrix1)

# Create a 2x3 matrix of random integers between 0 and 99
matrix2 = np.random.randint(0, 100, size=(2, 3))
print("\nMatrix 2:")
print(matrix2)

# Multiply the second matrix with the first modified matrix
result = np.dot(matrix2, matrix1)
print("\nResult:")
print(result)
```
