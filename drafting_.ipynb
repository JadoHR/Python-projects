```python
# function which return reverse of a string

def isPalindrome(s):
	return s == s[::-1]


# Driver code
s = input("Enter a five-digit integer or text: ")
ans = isPalindrome(s)

if ans:
	print("is a palindrome.")
else:
	print("is not a palindrome.")
```

```python
import math

def compute_pi(num_terms):
    pi_approx = 0
    for i in range(num_terms):
        pi_approx += (-1) ** i / (2 * i + 1)
    return 4 * pi_approx

def main():
    print("Terms\tApproximation")
    print("------------------------")

    target_values = [3.14, 3.141, 3.1415, 3.14159]
    target_terms = [0] * len(target_values)

    for num_terms in range(1, 101):  # Loop up to a maximum of 1000 terms
        pi_approx = compute_pi(num_terms)

        # Print the current approximation
        print(f"{num_terms}\t{pi_approx:.10f}")

        # Check if we've reached the target values
        for i, target in enumerate(target_values):
            if pi_approx >= target and target_terms[i] == 0:
                target_terms[i] = num_terms
                print(f"First time π >= {target}: {num_terms} terms")

    print("\nSummary:")
    for i, target in enumerate(target_values):
        print(f"π >= {target} after {target_terms[i]} terms")

if __name__ == "__main__":
    main()
```

```python
# Pattern 1: Left-aligned triangle
for i in range(1, 6):
    for j in range(i):
        print('*', end='')
    print()

# Pattern 2: Right-aligned triangle
for i in range(1, 6):
    spaces = ' ' * (5 - i)
    stars = '*' * i
    print(spaces + stars)

# Pattern 3: Inverted triangle
for i in range(5, 0, -1):
    stars = '*' * i
    print(stars)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

data = {'Hours': [2, 3, 4, 5, 6, 7, 8, 9, 10],
        'Scores': [30, 40, 50, 55, 60, 65, 70, 75, 80]}
```

```python
X = np.array(data['Hours']).reshape(-1, 1)
y = np.array(data['Scores'])
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X_new = np.array([11]).reshape(-1, 1)
predicted_score = model.predict(X_new)
print("Predicted Score:", predicted_score)
```

```python
import matplotlib.pyplot as plt

# Define the range of hours studied for the regression line
hours = np.linspace(2, 11, 100).reshape(-1, 1)

# Generate predictions for the corresponding range of hours
predicted_scores = model.predict(hours)

# Create a scatter plot of the actual data points
plt.scatter(data['Hours'], data['Scores'], color='green', label='Actual Data')

# Plot the regression line
plt.plot(hours, predicted_scores, color='yellow', label='Regression Line')

# Add labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.title('Simple Linear Regression')

# Display the plot
plt.legend()
plt.show()
```
