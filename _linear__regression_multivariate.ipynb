```python
from word2number import w2n
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv")
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df['experience'] = df['experience'].fillna('zero')
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
import math
median_test_score = math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
from sklearn import linear_model
# Separate features (x) and target
x = df.drop('salary($)', axis='columns')
target = df['salary($)']


# Assuming 'experience', 'test_score(out of 10)', and 'interview_score(out of 10)' are columns in your DataFrame
# Create and fit the linear regression model
reg = linear_model.LinearRegression()
reg.fit(x.values, target)

# Print coefficients and intercept
print('Coefficients:', reg.coef_)
print('Intercept:', reg.intercept_)
```

```python
print(df.columns)
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```

```python
import pandas as pd
from sklearn import linear_model

# Sample data
data = {'experience': [8, 8, 'five', 'two', 'seven', 'three', 'ten', 'eleven'],
        'test_score(out of 10)': [9, 6, 6, 10, 9, 7, None, 7],
        'interview_score(out of 10)': [5, 7, 10, 10, 6, 10, 7, 8],
        'salary($)': [50000, 45000, 60000, 65000, 70000, 62000, 72000, 80000]}

df = pd.DataFrame(data)

# Convert 'experience' column to numeric, handling errors with 'coerce' to replace non-numeric values with NaN
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

# Fill missing values with a chosen strategy, e.g., median
df = df.fillna(df.median())

# Separate features (x) and target
x = df.drop('salary($)', axis='columns')
target = df['salary($)']

# Create and fit the linear regression model
reg = linear_model.LinearRegression()
reg.fit(x.values, target)

# Print coefficients and intercept
print('Coefficients:', reg.coef_)
print('Intercept:', reg.intercept_)
```

```python
reg.predict([[12,10,10]])
```
