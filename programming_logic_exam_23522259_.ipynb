```python
# Melt the DataFrame to reshape it for plotting
df_melted = pd.melt(df, id_vars='country', var_name='Year', value_name='population')

# Convert 'Population' column to numeric, as it might be in string format
df_melted['population'] = pd.to_numeric(df_melted['population'], errors='coerce')
# Find the top x countries for each year
x = 7  # You can adjust this value based on how many top countries you want to display
top_countries_by_year = df_melted.groupby('Year').apply(lambda group: group.nlargest(x, 'population')).reset_index(drop=True)
plt.figure(figsize=(12, 6))

for year in top_countries_by_year['Year'].unique():
    # Filter out NaN values before plotting the histogram
    population_data = top_countries_by_year[(top_countries_by_year['Year'] == year) & ~top_countries_by_year['population'].isna()]['population']

    # Check if there are any non-NaN values before plotting
    if not population_data.empty:
        plt.hist(population_data, bins=35, alpha=0.6, label=year)

# Display the legend
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Histogram: Top Countries by Population Each Year')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

outcomes from the histogram shows opulation distribution of the top 7 or (x) countries for each year with frequency of every distribution (every data we have)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

def select_columns(df, selected_columns):
    return df[selected_columns + ['area']]

def build_linear_model(X, y):
    model = LinearRegression()
    model.fit(X, y)
    return model

def describe_linear_equation(model, feature_names):
    coefficients = list(zip(feature_names, model.coef_)) + [('Intercept', model.intercept_)]
    equation = " + ".join([f"{round(coeff, 4)}*{feat}" for feat, coeff in coefficients])
    return f"area = {equation}"

# Prepare dataset for simple linear regression (RH vs. Area)
simple_linear_columns = ['RH']
simple_linear = select_columns(forest_fires, simple_linear_columns)
X_simple = simple_linear[['RH']].values
y_simple = simple_linear['area'].values

# Build simple linear regression model
simple_linear_model = build_linear_model(X_simple, y_simple)
simple_linear_equation = describe_linear_equation(simple_linear_model, simple_linear_columns)
print("Simple Linear Equation:", simple_linear_equation)
```

```python
# Prepare dataset for multiple linear regression (RH, rain, wind, temp vs. Area)
multiple_linear_columns = ['RH', 'rain', 'wind', 'temp']
multiple_linear = select_columns(forest_fires, multiple_linear_columns)
X_multiple = multiple_linear[['RH', 'rain', 'wind', 'temp']].values
y_multiple = multiple_linear['area'].values

# Build multiple linear regression model
multiple_linear_model = build_linear_model(X_multiple, y_multiple)
multiple_linear_equation = describe_linear_equation(multiple_linear_model, multiple_linear_columns)
print("Multiple Linear Equation:", multiple_linear_equation)
```

Question number 3 for checking prime numbers from the given flowchart

```python
from math import isqrt

class PrimeChecker:
    def __init__(self, number):
        self.number = number
        self.is_prime = None

    def is_integer(self):
        return isinstance(self.number, int)

    def is_valid_number(self):
        return self.number > 1

    def check_prime(self):
        if not self.is_integer() or not self.is_valid_number():
            self.is_prime = "No"
            return self.is_prime

        i = 2
        while i <= isqrt(self.number):
            if self.number % i == 0:
                self.is_prime = "No"
                return self.is_prime
            i += 1
        self.is_prime = "Yes"
        return self.is_prime
```

```python
def main():
    while True:
        try:
            num = int(input("Write a number: "))
            prime_checker = PrimeChecker(num)
            if prime_checker.is_integer() and prime_checker.is_valid_number():
                if prime_checker.check_prime() == "Yes":
                    print(f"{num} True this is prime number.")
                else:
                    print(f"{num} False this is not a prime number.")
                break
            else:
                print("Invalid input. Please enter an integer greater than 1.")    # I kept error handling for non numerical vallues
        except ValueError:
            print("Error: Please enter a valid numerical value.")

if __name__ == "__main__":
    main()
```
